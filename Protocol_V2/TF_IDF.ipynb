```python
import pandas as pd
```

```python
df = pd.read_csv("queneau_fr.csv")
df.head()
```

```python
df["texte"]=df["texte"].str.lower()
df.head()
```

```python
from spacy.lang.fr.stop_words import STOP_WORDS
print(STOP_WORDS)
```

```python
for index , row in df.iterrows():

    df.loc[index, "texte"] = " ".join([word for word in row["texte"].split() if word not in STOP_WORDS])
    df.loc[index, "texte"] = " ".join([word for word in row["texte"].split() if len(word)>1])
    symbols = "!\"#$%&()*+-./:;<=>?@[\]^_`{|}~\n,"
    for i in symbols:
        df.loc[index, "texte"]=df.loc[index, "texte"].replace(i," ")
    df.loc[index, "texte"]=df.loc[index, "texte"].replace("'","")
    df.loc[index, "texte"]=df.loc[index, "texte"].replace("’","")

df.head()
```

```python
import spacy
nlp = spacy.load("fr_core_news_md")  # chargement du modéle dans Spacy
nlp.Defaults.stop_words |= STOP_WORDS
```

```python
for index , row in df.iterrows():
    doc = nlp(row["texte"])
    df.loc[index, "texte"] =" ".join([token.lemma_ for token in doc ])
```

```python
 from num2words import num2words
```

```python
for index , row in df.iterrows():
    text=""
    for word in row["texte"].split():
        if word.isnumeric():
            text=text+" "+num2words(word, lang='fr')
        else:
            text=text+" "+word
    df.loc[index, "texte"]=text
```

```python
df["DF"]=""
DF_TOTAL={}
for index,row in df.iterrows():
    DF={}
    for word in row["texte"].split():
        if word in DF.keys():
             DF[word]=DF[word]+1
        else:
            DF[word]=1
    DF_TOTAL.update(DF)
    row["DF"]=DF
```
