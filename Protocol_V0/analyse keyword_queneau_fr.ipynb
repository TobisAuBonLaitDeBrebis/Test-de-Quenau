```python
import pandas as pd
from collections import Counter
import plotly.graph_objects as go
```

```python
data = pd.read_csv(r"data/keyword_queneau_fr.csv")
data.head()
```

```python
def analyze_columns(column_name,title):
    list_of_keywords=list()
    for keyword_list in data[column_name]:
        keyword_list =keyword_list.replace('[','')
        keyword_list =keyword_list.replace(']','')
        keyword_list =keyword_list.replace("'",'')
        keyword_list =keyword_list.replace(' ,','')
        for keyword in keyword_list.split(","):
            list_of_keywords.append(keyword)

    list_of_unique_keywords_text=list(set(list_of_keywords))
    print("#############################################################")
    print("Nombre de mots clés générer: "+str(len(list_of_unique_keywords_text)))
    print("Nombre dérreure : "+str(len(data[data[column_name]=="error: min() arg is an empty sequence"])))
    print("#############################################################")
    print("\r")
    #print(set(list_of_keywords_text))

    collection_of_keywords=Counter(list_of_keywords_text)
    print(collection_of_keywords)
    x=list(collection_of_keywords.keys())
    y=list(collection_of_keywords.values())
    fig = go.Figure([go.Bar(x=x[:100], y=y[:100])])
    fig.update_layout(barmode='stack', xaxis={'categoryorder':'total descending'}, title=title)

    return list_of_keywords , fig
```

```python
list_of_keywords_text,fig_keywords_text =analyze_columns("keywords_text","Mots clés extrait à partir du texte complet")
fig_keywords_text.show(renderer="colab")
```

```python
list_of_keywords_text,fig_keywords_without_ent =analyze_columns("keywords_without_ent","Mots clés extraient à partir du texte sans les entités")
fig_keywords_without_ent.show(renderer="colab")
```

```python
list_of_keywords_only_VERB_ADJ_ADV_PROPN,fig_keywords_only_VERB_ADJ_ADV_PROPN =analyze_columns("keywords_only_VERB_ADJ_ADV_PROPN","Mots clés extraient  à partir des verbes, adjectifs, adverbe et nom propre")
fig_keywords_only_VERB_ADJ_ADV_PROPN.show(renderer="colab")

```

```python
list_of_keywords_only_VERBLemma_ADJ_ADV_PROPN,fig_keywords_only_VERBLemma_ADJ_ADV_PROPN =analyze_columns("keywords_only_VERBLemma_ADJ_ADV_PROPN","Mots clés extraient  à partir des verbes lemmatisé , adjectifs ,adverbe et nom propre")
fig_keywords_only_VERBLemma_ADJ_ADV_PROPN.show(renderer="colab")
```

```python
def figures_to_html(figs, filename="dashboard.html"):
    with open(filename, 'w') as dashboard:
        dashboard.write("<html><head></head><body>" + "\n")
        for fig in figs:
            inner_html = fig.to_html().split('<body>')[1].split('</body>')[0]
            dashboard.write(inner_html)
        dashboard.write("</body></html>" + "\n")


```

```python
tab_of_fig =[fig_keywords_text , fig_keywords_without_ent , fig_keywords_only_VERB_ADJ_ADV_PROPN , fig_keywords_only_VERBLemma_ADJ_ADV_PROPN]
```
